```python
files = [
    "GSE190940_P8NR.csv.gz",
    "GSE190940_P14NR.csv.gz",
    "GSE190940_P17NR.csv.gz",
    "GSE190940_P21NR.csv.gz",
    "GSE190940_P28NR.csv.gz",
    "GSE190940_P38NR.csv.gz",
    
    "GSE190940_P28DR.csv.gz",
    "GSE190940_P38DR.csv.gz",
    
    "GSE190940_P28DL.csv.gz",
] 
```

```python
genes = []
ddir = '../data/cheng21_cell_scrna/counts/'

for f in files:
    path = os.path.join(ddir, f)
    df = pd.read_csv(path, nrows=5)
    _genes = df.columns[1:].values
    if len(genes) > 0:
        assert np.all(genes == _genes)
    else:
        genes = _genes
        
    print(df.shape)
```

```python
unq, cnts = np.unique(genes, return_counts=True)
unq.shape, (cnts>1).sum()
```
